```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Reading the data from csv file into a pandas dataframe
df = pd.read_csv('SmartPhone_Sales_data.csv')
```

```python
#printing the entire dataframe
print(df)
```

```python
#plotting the data for analysis
for col in df.columns[1:]:
    plt.plot(df['Quarter'], df[col], label=col)

plt.xlabel('Quarter')
plt.ylabel('Number of smartphone sales')
plt.title('Quaterly smartphone sales')
plt.xticks(rotation = 40)
plt.legend()
plt.show()
```
